```python
from sklearn.cluster import KMeans
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
df = pd.read_csv('employee_data.csv')
df=df.head(1000)
df
```

```python
plt.scatter(df.age,df.salary)
```

```python
km = KMeans(n_clusters=3)
km
```

```python
ypredectid=km.fit_predict(df[['age','salary']])
ypredectid
```

```python
df['cluster']= ypredectid
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

df3
```

```python
plt.scatter(df1.age,df1.salary,color='green')
plt.scatter(df2.age,df2.salary,color='red')
plt.scatter(df3.age,df3.salary,color = 'blue')

plt.xlabel('age')
plt.ylabel('salary ($)')
plt.legend()
```

```python
sclaer = MinMaxScaler()
sclaer.fit(df[['salary']])
df['salary']=sclaer.transform(df[['salary']])
df
```

```python
sclaer.fit(df[['age']])
df['age']=sclaer.transform(df[['age']])
df
```

```python
km = KMeans(n_clusters=3)
ypredectid=km.fit_predict(df[['age','salary']])

df['cluster']= ypredectid
# df.drop(['id','groups'],axis='columns',inplace=True)
df


```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

df3

plt.scatter(df1.age,df1.salary,color='green')
plt.scatter(df2.age,df2.salary,color='red')
plt.scatter(df3.age,df3.salary,color = 'blue')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='orange',marker = '*',label='centroid')

plt.xlabel('age')
plt.ylabel('salary ($)')
plt.legend()
```

```python
km.cluster_centers_
```

```python
k_range=range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['age','salary']])
    sse.append(km.inertia_)
    
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of suaroid error')
plt.plot(k_range,sse)
```
